Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

1) Generate x values from -10 to 10 using np.linspace().

2)  Compute y values for:

o y= M⋅x
2

o y=M⋅sin(x)

3) Plot both functions in a single figure:

o Use different colors and line styles.

o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, 'r--', label=r'$y = M \cdot x^2$')
plt.plot(x, y2, 'b-', label=r'$y = M \cdot \sin(x)$')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title(f"Plots of y = M⋅x² and y = M⋅sin(x) for M = {M}")
plt.legend()
plt.grid(True)
plt.show()
```

Q.2 Create a dataset of five subjects and scores.

1) Convert it into a Pandas DataFrame.

2) Plot the scores using a Seaborn bar plot with:

o Different colors for each bar.

o Annotations on top of each bar.

o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Subjects": ["Math", "Physics", "Chemistry", "Biology", "Computer Science"],
    "Scores": [85, 78, 92, 88, 95]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
ax = sns.barplot(x="Subjects", y="Scores", hue="Subjects", data=df, palette="muted", legend=False)

for i, score in enumerate(df["Scores"]):
    ax.text(i, score + 1, str(score), ha='center', fontsize=12)

plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Scores in Different Subjects")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

1) Create a 2x2 subplot layout displaying:

o Line plot showing cumulative sum.

o Scatter plot with random noise.

2) Customize with titles, axis labels, and grids.


```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102317030
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(np.cumsum(data), 'b-', label="Cumulative Sum")
axes[0].set_title("Line Plot of Cumulative Sum")
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Cumulative Sum")
axes[0].grid(True)
axes[0].legend()

axes[1].scatter(range(50), data, c='r', marker='o', label="Random Noise")
axes[1].set_title("Scatter Plot with Random Noise")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Value")
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv

Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/company_sales_data.csv')

plt.figure(figsize=(10, 5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='b', label="Total Profit")
plt.title("Total Profit of All Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
for column in df.columns[1:-1]:
    sns.lineplot(x=df['month_number'], y=df[column], label=column)
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()

df.plot(kind="bar", figsize=(12, 6), width=0.8)
plt.title("Bar Chart of All Attributes")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
