```python
# 3.6 e) i. The average salary of employees.
#        ii. The total bonus paid to all employees.
#        iii. The youngest employee's age.
#        iv. The highest performance rating.

import pandas as pd
data = pd.read_csv("/content/employees.csv")

df = pd.DataFrame(data)

average_salary = df['Salary'].mean()

total_bonus = df['Bonus'].sum()

youngest_age = df['Age'].min()

highest_rating = df['Rating'].max()

print(f"i. Average Salary of Employees: {average_salary}")
print(f"ii. Total Bonus Paid to All Employees: {total_bonus}")
print(f"iii. Youngest Employee's Age: {youngest_age}")
print(f"iv. Highest Performance Rating: {highest_rating}")
```

```python
# 3.6 g) Add a new column that categorizes employees based on their performance rating:

#        i. Excellent for ratings >= 4.5
#        ii. Good for ratings >= 4.0 but < 4.5
#        iii. Average for ratings < 4.0

import pandas as pd
data = pd.read_csv("/content/employees.csv")

df = pd.DataFrame(data)

def categorize_performance(rating):
    if rating>=4.5:
       return 'Excellent'
    elif rating>=4.0 and rating<4.5:
       return 'Good'
    else:
       return 'Average'

df['Performance'] = df['Rating'].apply(categorize_performance)
print(df)
```
